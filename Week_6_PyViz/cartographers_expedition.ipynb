Don your cartographers hat and create map plots that will pinpoint the location of the places you'll be visitng. Use this as the map/intinerary for Harold's birthday expedition.

```python

import pandas as pd
pd.set_option('display.max_rows', None)
import plotly.express as px
import os
from pathlib import Path
from dotenv import load_dotenv
```

### Prep Mapbox API Key

```python
# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("MAP_BOX_API")

# Set the Mapbox API
px.set_mapbox_access_token(map_box_api)
```

### Prep Data

```python
# Read in data
places_of_interest = pd.read_csv('https://raw.githubusercontent.com/RutgersCodingBootcamp/RUT-SOM-VIRT-FIN-PT-08-2021-U-C/master/01-Lesson-Plans/06-PyViz/2/Activities/09-Stu_Cartographers_Expedition/Resources/harold_bday_plans.csv').dropna()
places_of_interest
```

### Route 1

```python
# Plot course for first route
first_route = places_of_interest[
    (places_of_interest["Name"].str.contains("Airport"))
    | (places_of_interest["Name"].str.contains("Central Park"))
]

route_1 = px.scatter_mapbox(first_route, lat="Latitude", lon="Longitude", color="Name")
route_1.show()
```

```python
fbprices['daily_pct_chg']=fbprices['price'].pct_change() #add column
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))   #setting canvas size
x=fbprices.index
y=fbprices['price'].values
plt.plot(x,y)
plt.title('FB Daily Price',fontsize=20)
```

```python
plt.figure(figsize=(20,10))   #setting canvas size
x=fbprices.index
y=fbprices['daily_pct_chg'].values
plt.plot(x,y)
plt.title('FB Daily % Change',fontsize=20)
```

```python
sorted_df=fbprices.sort_values(by='daily_pct_chg', ascending=False)
```

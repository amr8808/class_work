```python
import pandas as pd
%matplotlib inline
# from pathlib import Path
# csvpath = Path('../Resources/high_frequency_euro_ETF_bid_ask_spreads.csv')
df = pd.read_csv('high.csv', parse_dates=True, index_col='timestamp')
df.head()
```

```python
# Resample to 10 Seconds ('10S'), take the mean, and drop the NaN values
df2 = df.resample('10S').mean().dropna()
df2.head()
```

```python
# Plot the resulting bid-ask spread data
df2.bid_ask_spread.plot()
```

```python
# Calculate the autocorrelation
df2.bid_ask_spread.autocorr(lag=1)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Plot the ACF
plot_acf(df2['bid_ask_spread'], lags=30, zero=False) #armando what is this i forget
```

```python
# Plot the PACF
plot_pacf(df2['bid_ask_spread'], lags=30, zero=False)
```

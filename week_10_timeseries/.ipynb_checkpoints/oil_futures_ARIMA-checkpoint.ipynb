```python
import pandas as pd
%matplotlib inline
```

```python
# from pathlib import Path
# csvpath = Path('../Resources/oil_futures_front.csv')
df = pd.read_csv('TSLA.csv', parse_dates=True, index_col='Date')
df.head()
```

```python
df.Close.plot()
```

```python
# Differencing the Settle price stationarizes the time series
df.Close.diff().plot()
```

# Autocorrelation and Partial Autocorrelation Analysis

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# Plot the autocorrelation of the Settle Prices
plot_acf(df.Close, lags=20)
```

```python
# Plot the partial autocorrelation of the returns
plot_pacf(df.Close, lags=20)
```

# ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
# Create an ARIMA model
model = ARIMA(df.Close, order=(1, 1, 1))
```

```python
# Fit the model to the data
data = model.fit()
```

```python
# Print the model summary
data.summary()
```

```python
# Plot the Futures Price Forecast
pd.DataFrame(data.forecast(steps=10)[0]).plot(title="Futures Forecast")
```
